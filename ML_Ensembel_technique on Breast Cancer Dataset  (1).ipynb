### Ensemble Learning Techniques We Used to Build Model 

#### Bagging: Bagging is a technique where multiple models are trained on different subsets of the data and their predictions are combined to make the final prediction. We will explore the use of bagging with decision trees, random forests, and AdaBoost.

#### Boosting: Boosting is a technique where models are trained sequentially, with each subsequent model trying to correct the errors made by the previous model. We will explore the use of boosting with decision trees and AdaBoost.

#### We will evaluate the performance of each model using accuracy, precision, recall, F1-score, and area under the receiver operating characteristic curve (AUC-ROC).

# Label Encoding for target columns 

### _Label encoding is a technique used in machine learning to convert categorical data into numerical data. In many machine learning algorithms, numerical data is easier to work with than categorical data. Label encoding assigns a unique numerical value to each category in a categorical variable, allowing the data to be used in numerical computations. This technique is particularly useful when dealing with binary categorical variables, where only two categories are present_

```python
df['diagnosis'].value_counts()
```

```python
#replace no with 0
df['diagnosis']=df['diagnosis'].replace('B',0)
#replace yes with 1
df['diagnosis']=df['diagnosis'].replace('M',1)
```

```python
drop_list=['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
df1=df.drop(drop_list,axis=1)
df1.head()
```

### _Feature Selection with SelectKbest and Chi2_

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

x=df1.drop(['diagnosis'],axis=1)
y=pd.DataFrame(df1['diagnosis'])
select_feature=SelectKBest(chi2,k=5).fit(x,y)
X=select_feature.transform(x)
#selected feature are 'texture mean','area mean','concavity mean','area se','concavity worst'
X
```

# Bagging 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=10)
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

# Model_buliding

```python
from sklearn.ensemble import BaggingClassifier

from sklearn import tree

meta_estimator=BaggingClassifier(tree.DecisionTreeClassifier(random_state=10))

meta_estimator.fit(X_train,y_train)
```

```python
y_pred=meta_estimator.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
conf_matrix=pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actua:0','Actual:1'])
plt.figure(figsize=(8,5))
sns.heatmap(conf_matrix, annot=True,fmt='d', cmap='YlGnBu', cbar=False )
plt.show()
```

```python
from sklearn.metrics import classification_report
result=classification_report(y_test,y_pred)
print(result)
```

Here we got an accuracy of 0.94 % which is quite a good accuracy archived so but still, we try with other techniques also like Ada boost
or Xg boost to increase the accuracy score 
Bagging is a parallel technique and Boosting is a sequential technique we are successful in increasing the 0.010 % in the

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

plt.rcParams['figure.figsize']=(8,5)

fpr,tpr,threshold=roc_curve(y_test,y_pred)

#plot the roc curve 
plt.plot(fpr,tpr)

#set limit of x and y 
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])

#plt the straight line showing worst prediction for the model 
plt.plot ([0,1],[0,1],'r--')

#add the auc score
plt.text(x=0.05,y=0.8, s=('AUC Score:',round(roc_auc_score(y_test,y_pred),4)))

plt.xlabel('False positive rate (1-specificity)')
plt.ylabel('True positive rate (Sensitivity)')

plt.grid(True)
```

```python
from sklearn import metrics
# create the result table for all accuracy scores
# Accuracy measures considered for model comparision are 'Model', 'AUC Score', 'Precision Score', 'Recall Score','Accuracy Score','Kappa Score', 'f1 - score'

# create a list of column names
cols = ['Model', 'AUC Score', 'Precision Score', 'Recall Score','Accuracy Score','f1-score']

# creating an empty dataframe of the colums
result_tabulation = pd.DataFrame(columns = cols)

# compiling the required information
Bagging_Meta_estimator = pd.Series({'Model': "Bagging Meta-estimator",
                     'AUC Score' : metrics.roc_auc_score(y_test, y_pred),
                 'Precision Score': metrics.precision_score(y_test, y_pred),
                 'Recall Score': metrics.recall_score(y_test, y_pred),
                 'Accuracy Score': metrics.accuracy_score(y_test, y_pred),
                  'f1-score':metrics.f1_score(y_test, y_pred)})



# appending our result table
result_tabulation = result_tabulation.append(Bagging_Meta_estimator , ignore_index = True)

# view the result table
result_tabulation
```
